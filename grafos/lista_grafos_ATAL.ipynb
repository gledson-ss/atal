```python
class graph:
  def __init__(self):
    self._graphMap = {}
    self._listVertex = {}
    
  def addEdge(self, p1, p2):
    self._graphMap[p1].append(p2)
  
  def addVertex(self, v):
    self._listVertex[v] = True
    self._graphMap[v] = []

  def showEdges(self):
    for i in self._graphMap:
      for j in self._graphMap[i]:
        print('{} - {}'.format(i, j))
  
  def dfs(self, vx):
    vertex_visited = [False for i in range(len(self._graphMap))]
    stack = [vx]
    size_stack = 1
    while(size_stack):
      last = stack[-1]
      stack.pop()
      size_stack = size_stack - 1
      if(vertex_visited[last] == False):
        vertex_visited[last] = True
        print(last)

      for j in self._graphMap[last]:
        if(vertex_visited[j] == False):
          stack.append(j)
          size_stack = size_stack + 1
```

```python
g = graph()
```

```python
g.addVertex(0)
g.addVertex(1)
g.addVertex(2)
g.addVertex(3)
g.addVertex(4)
g.addVertex(5)
```

```python
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(1, 4)
g.addEdge(1, 5)
g.addEdge(2, 3)
g.addEdge(2, 5)
g.addEdge(3, 4)
g.addEdge(3, 5)
```

```python
g.showEdges()
```
